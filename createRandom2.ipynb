```python
import random
import pandas as pd
import numpy as np
import numpy.random as nr
from numpy import *
```

```python
#データセット①
x1 = 10 #
y1 = 20 #データ幅の指定
zz = 3000 #出力データ個数(固定)
r1 = 2 #出力データの小数点指定(整数は０)
```

```python
#出力モデル
a = np.random.uniform(x1, y1, zz)
a = np.round(a, r1)
a
```

```python
#データセット②
x2 = 0.005
y2 = 0.1 #データ幅の指定
r2 = 4 #出力データの小数点指定(整数は０)
```

```python
b = np.random.uniform(x2, y2, zz)
b = np.round(b, r2)
b
```

```python
#データセット③
x3 = 1000
y3 = 2000 #データ幅の指定
r3 = 3 #出力データの小数点指定(整数は０)
```

```python
c = np.random.uniform(x3, y3, zz)
c = np.round(c, r3)
c
```

```python
#データセット④
x4 = 0
y4 = 1 #データ幅の指定
r4 = 0 #出力データの小数点指定(整数は０)
```

```python
d = np.random.uniform(x4, y4, zz)
d = np.round(d, r4)
d
```

```python
#時間設定
#開始時間,freqで時間幅を設定
#リファレンス http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases
dt = pd.date_range('2014-11-01 10:00', periods=zz, freq='2H30T17S')
dt
```

```python
#データセット⑤
wl = np.array(["パン","牛乳","お肉","お水","砂糖","お茶","野菜","菓子","果物","お米"])
wd = random.choice(wl, zz)
wd
```

```python
#index
index = np.arange(0,zz,1)
index
```

```python
#それぞれを統合
#各カラムの名前を設定
df = pd.DataFrame(
{'文字列' : wd,
 '時間' : dt,
 'センサー１' : a,
 'センサー２' : b,
 'センサー３' : c,
 'センサー４' : d})
df
```

```python
#CSVで出力
df.to_csv('output.csv', index=False)
```
