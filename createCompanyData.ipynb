```python
#ここはとりあえずSHIFT+ENTERでスキップ
import random
import pandas as pd
import numpy as np
import numpy.random as nr
from numpy import *
```

```python
#サンプル数定義：作成するデータの件数を定義します
dataCount = 3000
```

```python
#データセットサンプル：ランダムな整数
#最小値
x1 = 500 
#最大値
y1 = 1000 
#出力データの小数点指定(整数は０)
r1 = 0 
#列を作成
a = np.random.uniform(x1, y1, dataCount)
a = np.round(a, r1)
a
```

```python
#データセットサンプル;文字列
#作成させたい単語リストを配列形式で作成
wl = np.array(["パン","牛乳","お肉","お水","砂糖","お茶","野菜","菓子","果物","お米"])
#列を作成
wd = random.choice(wl, dataCount)
wd
```

```python
#データセットサンプル：ランダムな時間データ
#開始時間,freqで時間幅を設定
#リファレンス http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases
dt = pd.date_range('2014-11-01 10:00', periods=dataCount, freq='2H30T17S')
dt
```

```python
#データセットサンプル：INDEXデータ
#０から１ずつ足していく
index = np.arange(0,dataCount,1)
index
```

```python
#ESC+Aでセルを追加できるのでここから以下でデータを定義
```

```python
#データセットサンプル：ランダムな整数
#最小値
x1 = 10
#最大値
y1 = 20
#出力データの小数点指定(整数は０)
r1 = 0 
#列を作成
b = np.random.uniform(x1, y1, dataCount)
b = np.round(b, r1)
b
```

```python
#ここまででデータを作成
```

```python
#最後に列を統合してCSV化する
#各カラムの名前を設定　'列名' : wd(列の変数名),でつないでいく
df = pd.DataFrame(
{'品物' : wd,
 '値段' : a,
 '販売数' : b})
```

```python
df
```

```python
#CSVで出力
df.to_csv('output.csv', index=False)
```
